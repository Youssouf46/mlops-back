```python
plt.figure(figsize=(14,5))
plt.title('away_score Distribution FIFA 20')
sns.distplot(a=df['away_score'], kde=True, bins=20)
plt.axvline(x=np.mean(df['away_score']),c='orange',label='Mean away_score of All Players')
plt.legend()
```

```python
plt.figure(figsize=(14,5))
plt.title('away_score Distribution FIFA 20')
sns.distplot(a=df['home_score'], kde=True, bins=20)
plt.axvline(x=np.mean(df['home_score']),c='orange',label='Mean away_score of All Players')
plt.legend()
```
